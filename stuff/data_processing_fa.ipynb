# McDonalds and Starbucks: what would be better as meals for teens?
<b>1. Data processing & exploratory analysis</b>

DONE:
- Norms values from official document-1 extracted manually --> txt --> dict. We find only 6 values: <b>Energy(kcal), Fat(g), Carbohydrates(g), Fiber(g), Protein(g), Sodium (g)</b>, wich exist in all three datasets (document-1, McD and SB menus).
- Values from McD and SB .csv menus --> Excel --> (processing: combine, add features in cols).
- Features added: (1) <b>'McD/SB'</b>: restaurant coding, (2) <b>'Category'</b>: food/drink, (3) <b>'Kind'</b>: certain item group, for instance: 'bakery', 'hot breakfast', 'sweets', 'hot drink'. SB have about 2 times more such item groups than McD but the latter have more items per groups.
- Dataset enriched manually: >60 items in <b>'Snacks & sweets'</b> category were added for SB menu. Thus we have more balanced dataaset (McD/SB share).
- Discard from decimal zeroes: dividing all mass values by 1000 (mg --> g).
- Several bug fixes in values.
- Deleted items with all zero values (for instance, <b>Iced Coffee</b>).
- Calculate weighted norms - according the rule requirement from official document-2: <i>"...for children meals breakfast / dinner share for caloric and nutrition values should be about 25% / 35% of daily total"</i>. We set 30% / 30% share --> norm for breakfast / dinner =  <b>norms*0.3</b> (only breakfast and dinner were mentioned in the task).
- Deleting items <b>with > 60%</b> (breakfast + dinner total: 'sum_norm' dict) of daily norm caloric or nutrition values (for instance, <b>1880 kcal 40 pieces of chicken McNuggets</b>). Exception: we assume that sodium needs increases due to sport activity: the <b>norm value 0.78 g</b> were update to <b>1.0 g</b> sodium threshold. It allows preserve more items in dataset.
- Exploratory data analysis and visualizations: total counts, values means, several insights.

```python
# required imports
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.dpi"] = 144
%matplotlib inline
```

### Official norms

```python
# Dict with actual norms (doc-1)
norms = {}

f = open ('actual_norms.txt')

for line in f:
    line = line.strip().split(',')
    norms[line[0]] = line[1]
    
del norms['Item']
norms
```

```python
# calculating weighted norms:
meal_norm = {key:float(value)*0.3 for key, value in norms.items()}
    
# sum_norm dict for data filtering (del items with greater values):   
sum_norm = {key:float(value)*0.6 for key, value in norms.items()}
    
print (meal_norm)
print (sum_norm)
```

### McD, SB menus

```python
# Starbucks (SB) and McDonalds (McD) combined menu in single df: 
menu_df = pd.read_csv ('combined.csv', sep=';', encoding = 'koi8-r')
menu_df.head()
```

```python
# checking data gaps - it's OK
menu_df.describe()
```

### Clean the data

```python
# any all-zero items to delete?
del_list = []
def find_zeroes (row):
    
    '''the function recieve row of a df and process it: if all quantitative fields 
    are zeroes in the row, the row included in list for deleting which no returned directly'''
    
    if row[4:].sum() == 0:
#         print (row)
        del_list.append (row[3])
        
menu_df.apply (find_zeroes, axis = 1)
menu_df = menu_df.loc[~menu_df['Item'].isin(del_list)]
len (menu_df) # 7 values deleted
```

```python
# filtering: del items with values > weighted norm in sum_norm dict:
menu_df[menu_df['Energy(kcal)']>=sum_norm['Energy(kcal)']]
# only 1 item exceeds sum_norms (1740 kcal)
```

```python
menu_df[menu_df['Fat(g)']>=sum_norm['Fat(g)']]
# 3 items exceed sum_norms (58.2 g)
```

```python
menu_df[menu_df['Carbohydrates(g)']>=sum_norm['Carbohydrates(g)']]
# no items exceed sum_norms
```

```python
 
# 2 items exceed sum_norms (12.0 g)
```

```python
menu_df[menu_df['Protein(g)']>=sum_norm['Protein(g)']]
# only 1 item exceeds sum_norms (52.2 g)
```

```python
menu_df[menu_df['Sodium(g)']>=sum_norm['Sodium(g)']]
# Wow! 104 items exceed sum_norms (0.78 g)
```

```python
#menu_df(menu_df[menu_df['Sodium(g)']>=sum_norm['Sodium(g)']])['McD/SB'].count()
menu_df[menu_df['Sodium(g)']>=sum_norm['Sodium(g)']]['McD/SB'].count()
```

```python
menu_df[(menu_df['Sodium(g)']>=sum_norm['Sodium(g)'])&(menu_df['McD/SB']=='SB')]['McD/SB'].count()
```

```python
menu_df[(menu_df['Sodium(g)']>=sum_norm['Sodium(g)'])&(menu_df['McD/SB']=='McD')]['McD/SB'].count()
```

```python
# assumption: sodium needs increases due to sport activity: 
# 1.0 g sodium threshold. Nevertheless 73 items exceed 1.0 g
menu_df[menu_df['Sodium(g)']>=1]
```

```python
menu_df[(menu_df['Sodium(g)']>=1)&(menu_df['McD/SB']=='SB')].count()
```

```python
# Drop out all items that exceeds the criteria:
menu_df = menu_df[menu_df['Energy(kcal)'] < sum_norm['Energy(kcal)']]
menu_df = menu_df[menu_df['Fat(g)'] < sum_norm['Fat(g)']]
menu_df = menu_df[menu_df['Fiber(g)'] < sum_norm['Fiber(g)']]
menu_df = menu_df[menu_df['Protein(g)'] < sum_norm['Protein(g)']]
menu_df = menu_df[menu_df['Sodium(g)'] < 1.0]

len (menu_df)
```

```python
# We increased sodium sum norm 0.78 --> 1.0, i.e. +0.22. 
# In turns it increases sodium meal_norm by 0.11 g. Dicts update: 
meal_norm['Sodium(g)'] += 0.11
sum_norm['Sodium(g)'] += 0.22
print (meal_norm)
print (sum_norm)
```

```python
# writing processed dataset in csv
menu_df.to_csv ('combined_processed.csv', index=False)
```
