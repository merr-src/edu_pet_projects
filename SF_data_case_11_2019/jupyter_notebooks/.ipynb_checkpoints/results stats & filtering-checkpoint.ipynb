# McDonalds and Starbucks: what would be better as meals for teens?
<b>3. Results stats and cocnclusions</b>

0. Open all result datasets as df's via file system (<b>os</b> lib) and concatenate them to general <b>'mega_data'</b> df.
1. Checking several <b>MEAN & STDEV scenarios</b>: filtering best-balanced meal combinations (MEAN is 1-st, STDEV is 2-nd criterion):

- MEAN = 1<b>+/- 0.1</b>, STDEV <b><= 0.1</b>
- MEAN = 1<b>+/- 0.25</b>, STDEV <b><= 0.2</b>
- MEAN = 1.2<b>+/- 0.1</b>, STDEV <b><= 0.1</b> ('power meal' for sports)
- ...

2. Evaluation of <b>SB VS McD</b> menus processed the filter. 
3. Draw the conclusions. Preference for <b>hot</b> dishes and <b>moderate-sugared</b> drinks and meals

```python
# required imports
import os
import pandas as pd
import matplotlib
%matplotlib inline
```

```python
# loop for building all-results df
# 3-item dfs have 13 cols (1-st food, 2-nd food, 3-rd drink) whereas 
# 2-item dfs have 12 cols (1-st food, 2-nd drink). Adding 13-th 'filler' 
# with '0' for all dfs for proper concatenation

files = os.listdir('results')

mega_data = pd.DataFrame()

for file in files:
    temp = pd.read_csv(os.path.join('results', file))
    if len(temp.columns) == 12:
        temp.insert(4, 'third', 0)
        
    mega_data = pd.concat([mega_data, temp], sort=False)
    
mega_data # 360774 lines!
```

```python
# Filtering best-balanced combinations:
# (1) MEAN should be ≈ 1. Let's assume +/- 0.1 as criterion
mega_data_best = mega_data[(mega_data['MEAN'] > 0.9) & (mega_data['MEAN'] < 1.1)]
mega_data_best # 360774 lines --> 74416 lines: only 20% pass the filter
```

```python
# Filtering best-balanced combinations:
# (2) STDEV should be as less as possible. Let's assume <= 0.1 as criterion
mega_data_best_2 = mega_data_best[mega_data_best['STDEV'] <= 0.1]
mega_data_best_2 # 74416 lines --> 521 lines only 0.7% pass second filter
```

```python
counts = mega_data_best_2.groupby (['McD/SB']).count()
counts ['Kind']
```

298 McD's best combinations VS 223 SB --> <b>McD overall have about one third more best combinations</b>. 

If only 'solid' 3-item meal taking into account: 218 McD's best combinations VS same 223 SB --> <b>SB breakfasts & dinners have slight preponderance</b>. 

Interestingly that <b>all SB light breakfasts and dinners as snacks cannot pass the filter</b>.

```python
counts['Kind'].plot(kind = 'bar', subplots = False, width = 0.5, figsize = (6,3))
```

### What if... not so strong criteria?

```python
# (1) mean should be ≈ 1. Let's assume +/- 0.25 as criterion
mega_data_good = mega_data[(mega_data['MEAN'] > 0.75) & (mega_data['MEAN'] < 1.25)]
mega_data_good # 360774 lines --> 179758 lines: 50% pass the filter
```

```python
# (2) Std should be as less as possible. Let's assume <= 0.2 as criterion
mega_data_good_2 = mega_data_best[mega_data_best['STDEV'] <= 0.2]
mega_data_good_2 # # 179758 lines --> 5538 lines: 3% pass second filter
```

```python
counts_2 = mega_data_good_2.groupby (['McD/SB']).count()
counts_2 ['Kind']
```

2602 McD's best combinations VS 2936 SB --> <b>SB overall have about 12% more best combinations</b> 

If only 'solid' 3-item meal taking into account:1519 McD's best combinations VS same 2925 SB --> <b>SB breakfasts & dinner shave little less than two-fold preponderance</b>

In this case SB dinners significantly increased as main SB best group. Also SB breakfasts group accounts for the result. <b>SB light breakfasts and dinners and snacks have no weight as in the first case.</b>

```python
counts_2['Kind'].plot(kind = 'bar', subplots = False, width = 0.5, figsize = (6,4))
```

### 'Power meal' for sports

```python
# +20% of norm for calories and nutritions for sport activity:
# (1) mean should be ≈ 1. Let's assume +/- 0.1 as criterion
mega_data_sport = mega_data[(mega_data['MEAN'] > 1.1) & (mega_data['MEAN'] < 1.3)]
mega_data_sport # 360774 lines --> 63381 lines: 17.5% pass the filter
```

```python
# (2) Std should be as less as possible. Let's assume <= 0.2 as criterion
mega_data_sport_2 = mega_data_sport[mega_data_sport['STDEV'] <= 0.2]
mega_data_sport_2 # 179758 lines --> 2010 lines: 3% pass second filter
```

```python
#  Due to sports activity is reasonable to figure an additional filter: Energy >= 1.2
# (viewing the results shows that many items could be insufficient on Calories)
mega_data_sport_3 = mega_data_sport_2[mega_data_sport_2['Energy'] >= 1.2]
mega_data_sport_3 # 2010 lines --> 1002 lines: 50% pass third filter
```

```python
counts_3 = mega_data_sport_3.groupby (['McD/SB']).count()
counts_3 ['Kind']
```

<b>Total prevalence of McD meals</b> (which is expected): 996 VS 36. 

### Conclusions and meal recommendations: McD or SB?

<b>1. Summarize all that we revealed: exploratory analysis insights:</b>

 
- SB have much more food items (<b>136 in 217 total</b>), whereas McD much more drinks (<b>138 in 201 total</b>)

- Top on <b>Calories</b> (and also in carbohydrates, likely sugar) food group: McD sweets/snacks (McFlurry ice-creams). Other high-energy groups: SB Cold sandwiches (also leader on Fat), SB Protein boxes (also leader on Fat), McD Hot big breakfast. Mean ≈ median <b>Calories</b> and <b>Carbohydrates</b> for both McD and SB all-food and all-drinks groups. McD items much more caloric-rich and rich of <b>Carbohydrates</b> in average. 

- Also top on <b>Fat</b> is McD Sweets (ice-creams) and Burgers. All-food and all-drinks mean ≈ median just for food. Outliers on <b>Fat</b> value in both SB and McD drinks are expected. In average McD food contains much more <b>Fat</b>.

- Top on <b>Protein</b>: SB cold sandwiches (also leaders in Fiber). Also high-protein is SB Protein boxes and Warm sandwiches, as McD Burgers and McD Salads. McD mean slightly < median, for SB vice versa. Seems that SB food have more outliers - items with relative high <b>Protein</b> contains (or groups with protein-rich items). Nevertheless, McD overwight in absolute mean & median values.

- Also top on <b>Fiber</b>: SB salads. Mean ≈ median just for food (with zero value in drinks, as expected). SB food ≈ McD food in average <b>Fiber</b>.  

- Top <b>Sodium</b>: SB Cold sandwiches, McD burgers and McD Hot big breakfast. Same situation as with <b>Protein</b> for all-food and also for all-drinks. In average McD food contains much more <b>Sodium</b>.

McD food and drinks exceed SB ones on average <b>Calories, Fat, Carbohydrates, Protein and Sodium</b>. On this values <b>McD : SB ≈ 60 : 40</b>, whereas <b>fiber</b> ratio close to <b>50 : 50</b>. It is being more prominent taking into account that McD menu have much more drinks VS SB menu.

<b>2. Summarize menu selection scenarions completed</b>

Due to combinatorial algorithm based on SB and McD menus we obtain over 360K samples of 3-item and 2-item breakfast / dinner. The next step is selection the best of them with weighted 6-value totals. MEAN and STDEV of the totals have choosen as criteria to evaluate several scenarios. MEAN = 1 refers to norm value, MEAN = 1.2 refers to norm value*1.2, i.e. norm +20% e t.c.

- <b>MEAN = 1+/-0.1, STDEV <= 0.1</b> ('tight-norm', scenario): 360K --> 571 best samples passed the filter: 298 McD VS 223 SB --> overall <b>McD have about one third more best combinations</b>. If only 3-item meals taking into account: 218 McD VS 223 SB --> <b>SB breakfasts & dinners have slight preponderance</b>. All SB light breakfasts, dinners and snacks cannot pass the filter.


- <b>MEAN = 1+/-0.25, STDEV <= 0.2</b> ('mild-norm', scenario). 360K --> 5538 best samples passed the filter: 2602 McD VS 2936 SB --> overall <b>SB have about 12% more best combinations</b>. If only 3-item meal taking into account: 1519 McD's best combinations VS same 2925 SB --> <b>SB breakfasts & dinners have about two-fold preponderance</b>. SB dinners significantly increased as main SB best group. Also SB breakfasts group accounts for the result. SB light breakfasts, dinners and snacks have no weight as in the first case.


- <b>MEAN = 1.2+/-0.1, STDEV <= 0.1, Energy >= 1.2 norm</b> ('power meal for sports' scenario). 360K --> 1002 best samples passed the filter: 996 McD VS 36 SB. <b>Total prevalence of McD meals</b> (which is expected due to more caloric-rich nutrient-saturated food).

<b>3. What restaurant better?</b>
Generally the answer depends on the aim. Our analysis of three scenarios shows that even client focused on balanced meals that fitted norms they should vote for McD menu. If norm values are not so important (but slill accounted) - SB menu will be much more preferrable. Finally when one need high-caloric and nutrition-rich meal due to spotrs, activities e t.c. they should choose McD menu.

<b>NOTE:</b> final 'best combinations' obtained is just filtered on formal criteria. We hope that this lists is enough to provide <b>extra variations</b>, from which one can choose any <b>items that completely fitted their needs</b> (for example account of allergics). What could be also accounted?

- <b>Hot items preference:</b> warm sandwiches instead of cold sandwiches in SB menu, hot drinks instead of cold drinks in both menus. Idea is that child meal cannot be absolutely cold.
- <b>High-sugar warning:</b> one could make discretion especially about McD very high-caloric and sugar-rich ice-creams.
- <b>Salty food warning:</b> we allready increase the sodium norm value from 0.78 to 1.0 g per meal as this value allow to preserve much more items in original combined menu. We bellieve that it is acceptable since sodium consumption increase due to sports activity. Nevertheless we recommend to care about it and provide additional drinking water on meals and during sports and activities.

### Shortcomings of the methodology:
1. Average norms from orginal values for girls and boys. As values for <b>Energy</b>, <b>Fat</b>, <b>Carbohydrates</b> and <b>Protein</b> for boys is bigger (on about 16%) averaged values can cause insufficient resulting values for boys (-8%) and excessive ones for girls (+8%). One can take it into account and perform additional filter scenarios for girls or boys. Values for <b>Fiber</b> and <b>Sodium</b> are equal for both sexes.
2. As stated in case description we should use original McD and SB datasets from client... So these datasets based on USA McD and SB menus whereas our task is in the Russian context (we interpreted it so and search for Russian official dietary documents).
3. Also we revealed that menu datasets from client were partly outdated (this is another reason to add >60 items in SB menu - it allready figured on website but absent in the data).
