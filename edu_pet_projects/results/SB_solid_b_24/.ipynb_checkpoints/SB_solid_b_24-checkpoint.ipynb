## Analytics in terms of B-index

### 1. "Solid breakfast" SB combination set 

```python
# required imports
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# list with 24 combinations of 3 items, they specified in same order 
# as B_index cols in df below: 1-st line '('Soups', 'Bakery', 'Cold Drinks')' 
# --> 1-st col 'SB_solid_b_0' and so on
combinations = pd.read_csv ("SB_solid_b_24.txt", names = ['1','2','3'])
combinations
```

```python
SB_solid_b_24 = pd.read_csv ('SB_solid_b_24.csv', sep=(','))
SB_solid_b_24.head ()
```

```python
# different combinations of groups gave different combinations of items
SB_solid_b_24.info()
```

```python
# cols of different length!
SB_solid_b_24.describe()
```

```python
# mean B-index of the subset:
SB_solid_b_24.mean().value_counts (bins=7)
```

```python
# B_index mean values for all groups
cols = pd.Index.to_list(SB_solid_b_24.columns[0:])
SB_solid_b_24[cols].mean().plot(kind = 'bar', 
                                subplots = False, 
                                width = 0.5, 
                                figsize = (17,5),
                                grid = True)
```
