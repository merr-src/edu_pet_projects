## Analytics in terms of B-index

### 1. "Solid dinner" SB combination set 

```python
# required imports
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# list with 24 combinations of 3 items, they specified in same order 
# as B_index cols in df below --> 1-st col 'SB_solid_b_0' and so on

combi_list = []

with open ('SB_solid_d_24.txt', 'r') as f:
    for line in f:
        line = line[:-1]
        combi_list.append (line)
    f.close()
```

```python
SB_solid_d_24 = pd.read_csv ('SB_solid_d_24.csv',  skiprows=1, names=combi_list , sep=(','))
SB_solid_d_24.head()
```

```python
# cols of different length!
SB_solid_d_24.describe()
```

```python
# B_index mean values for all groups
cols = pd.Index.to_list(SB_solid_d_24.columns[0:])
SB_solid_d_24[cols].mean().plot(kind = 'bar', subplots = False, width = 0.5, figsize = (17,5), grid = True)
```

```python
# delete all groups which not in (0.75 - 1.25) means interval:
# (why simple df[df['name'].mean < N)] do not work in df with different column lengths?)

# bellow cutoff:
bellow_cut = (SB_solid_d_24.mean() < 0.75)
bellow_drop_list = [col for col in bellow_cut]
b = bellow_cut.loc[bellow_drop_list]
bellow_drop = pd.Index.to_list(b.index)

# above cutoff:
above_cut = (SB_solid_d_24.mean() > 1.25)
above_drop_list = [col for col in above_cut]
a = above_cut.loc[above_drop_list]
above_drop = pd.Index.to_list(a.index)

# deleting columns:
SB_solid_d_24.drop (columns = bellow_drop, inplace=True)
SB_solid_d_24.drop (columns = above_drop, inplace=True)
SB_solid_d_24.describe()
```

```python
# first item - 'Cold Sandwiches' hence it is argument for preserving only 
# combinations with 'Hot Drinks':
cols = pd.Index.to_list(SB_solid_d_24.columns)

SB_solid_d_24.drop (columns = [col for col in cols if 'Cold Drinks' in col], inplace=True)

# vizualizing filtered df:
cols = pd.Index.to_list(SB_solid_d_24.columns[0:])
SB_solid_d_24[cols].mean().plot(kind = 'bar', subplots = False, width = 0.5, figsize = (6,4), grid = True)
```
