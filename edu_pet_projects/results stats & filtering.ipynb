### Results filtering & stats 

```python
# required imports
import os
import pandas as pd
import matplotlib
%matplotlib inline
```

```python
files = os.listdir('results')
# all_best = pd.read_csv(os.path.join('results', files[0]))
# all_best
```

```python
# loop for building all-results df
# 3-item dfs have 13 cols (1-st food, 2-nd food, 3-rd drink) whereas 
# 2-item dfs have 12 cols (1-st food, 2-nd drink). Adding 13-th 'filler' 
# with '0' for all dfs for proper concatenation

mega_data = pd.DataFrame()

for file in files:
    temp = pd.read_csv(os.path.join('results', file))
    if len(temp.columns) == 12:
        temp.insert(4, 'third', 0)
        
    mega_data = pd.concat([mega_data, temp], sort=False)
    
mega_data # 360774 lines!
```

```python
# Filtering best-balanced combinations:
# (1) mean should be ≈ 1. Let's assume +/- 0.1 as criterion
mega_data_best = mega_data[(mega_data['MEAN'] > 0.9) & (mega_data['MEAN'] < 1.1)]
mega_data_best # 360774 lines --> 74416 lines: only 20% pass the filter
```

```python
# Filtering best-balanced combinations:
# (2) Std should be as less as possible. Let's assume <= 0.1 as criterion
mega_data_best_2 = mega_data_best[mega_data_best['STDEV'] <= 0.1]
mega_data_best_2 # 74416 lines --> 521 lines only 0.7% of this 20% pass second filter
```

```python
counts = mega_data_best_2.groupby (['McD/SB']).count()
counts ['Kind']
# 298 McD's best combinations VS 223 SB --> McD overall have about 
# one third more best combinations 

# If only 'solid' 3-item meal taking into account:
# 218 McD's best combinations VS same 223 SB --> SB breakfasts & dinners
# have slight preponderance

# Interestingly that all SB light breakfasts and dinners as snacks
# can not pass the filter
```

```python
counts['Kind'].plot(kind = 'bar', subplots = False, width = 0.5, figsize = (6,4))
```

### What if... not so strong criteria?

```python
# (1) mean should be ≈ 1. Let's assume +/- 0.25 as criterion
mega_data_good = mega_data[(mega_data['MEAN'] > 0.75) & (mega_data['MEAN'] < 1.25)]
mega_data_good # 360774 lines --> 179758 lines: 50% pass the filter
```

```python
# (2) Std should be as less as possible. Let's assume <= 0.2 as criterion
mega_data_good_2 = mega_data_best[mega_data_best['STDEV'] <= 0.2]
mega_data_good_2 # # 179758 lines --> 5538 lines: 3% of that 50% pass second filter
```

```python
counts_2 = mega_data_good_2.groupby (['McD/SB']).count()
counts_2 ['Kind']
# 2602 McD's best combinations VS 2936 SB --> SB overall have about 
# 12% more best combinations 

# If only 'solid' 3-item meal taking into account:
# 1519 McD's best combinations VS same 2925 SB --> SB breakfasts & dinners
# have little less than two-fold preponderance

# In this case SB dinners significantly increased as main SB best group
# also SB breakfasts group accounts for the result
# SB light breakfasts and dinners and snacks have no weight as in the first case
```

```python
counts_2['Kind'].plot(kind = 'bar', subplots = False, width = 0.5, figsize = (6,4))
```
